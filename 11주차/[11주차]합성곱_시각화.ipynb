```python
from tensorflow import keras
```

```python
# fasion mnist
# CNN 모델 만들고 학습해서
# 레이어 추출
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# CNN모델 만들기
# C - P - D - F - D - D
```

```python
x_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(32,kernel_size=3,activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPool2D(),
    keras.layers.Dropout(0.3),
    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
```

```python
model.summary()

# Conv2D 레이어의 출력
# output_height = (input_height - kernel_size + 2 * padding) / stride + 1
# output_width = (input_width - kernel_size + 2 * padding) / stride + 1
# output_channels = filters
# output_height = (28 - 3 + 2 * 0) / 1 + 1 = 26
# output_width = (28 - 3 + 2 * 0) / 1 + 1 = 26

# MaxPooling2D 레이어의 크기는 (2, 2). 따라서 출력 높이와 너비는
# output_height = (input_height - pool_size) / strides + 1
# output_width = (input_width - pool_size) / strides + 1
# output_height = (26 - 2) / 2 + 1 = 13
# output_width = (26 - 2) / 2 + 1 = 13
```

```python
# 모델 컴파일
model.compile(optimizer='adam',loss=keras.losses.SparseCategoricalCrossentropy(), metrics='accuracy')
history = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```
